```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
process = pd.read_csv('stress_test_log_process.csv')
thread = pd.read_csv('stress_test_log_thread.csv')
```

```python
# Change pallete
sns.set_palette("pastel")
```

```python
# Assume df is your DataFrame
sns.set_style("white")
plt.figure(figsize=(10, 6))
sns.heatmap(process.drop([
    'Nomor',
    'Jumlah worker client yang gagal', 
    'Jumlah worker server yang gagal',
    'Jumlah worker client yang sukses',
    'Jumlah worker server yang sukses'
    ],axis=1).corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Korelasi Antar Variabel")
plt.xticks(rotation=45)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 8))

# Process data plot
sns.lineplot(data=process, x='Waktu total per client', y='Throughput per client', 
             hue='Volume', markers=True, dashes=False,
             palette='crest', ax=axes[0])
axes[0].set_title("Process: Throughput vs Time")

# Thread data plot
sns.lineplot(data=thread, x='Waktu total per client', y='Throughput per client', 
             hue='Volume', markers=True, dashes=False,
             palette='crest', ax=axes[1])
axes[1].set_title("Thread: Throughput vs Time")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

sns.lineplot(data=process, x='Jumlah client worker pool', y='Throughput per client', hue='Volume', ax=axes[0,1], palette='crest')
axes[0,1].set_title("Process: Throughput vs Client Workers")

sns.lineplot(data=process, x='Jumlah server worker pool', y='Throughput per client', hue='Volume', ax=axes[0,0], palette='crest')
axes[0,0].set_title("Process: Throughput vs Server Workers")

sns.lineplot(data=thread, x='Jumlah client worker pool', y='Throughput per client', hue='Volume', ax=axes[1,1], palette='crest')
axes[1,1].set_title("Thread: Throughput vs Client Workers")

sns.lineplot(data=thread, x='Jumlah server worker pool', y='Throughput per client', hue='Volume', ax=axes[1,0], palette='crest')
axes[1,0].set_title("Thread: Throughput vs Server Workers")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

sns.lineplot(data=process, x='Jumlah client worker pool', y='Waktu total per client', hue='Volume', ax=axes[0,0], palette='crest')
axes[0,0].set_title("Process: Total Waktu vs Client Workers")

sns.lineplot(data=process, x='Jumlah server worker pool', y='Waktu total per client', hue='Volume', ax=axes[0,1], palette='crest')
axes[0,1].set_title("Process: Total Waktu vs Server Workers")

sns.lineplot(data=thread, x='Jumlah client worker pool', y='Waktu total per client', hue='Volume', ax=axes[1,0], palette='crest')
axes[1,0].set_title("Thread: Total Waktu vs Client Workers")

sns.lineplot(data=thread, x='Jumlah server worker pool', y='Waktu total per client', hue='Volume', ax=axes[1,1], palette='crest')
axes[1,1].set_title("Thread: Total Waktu vs Server Workers")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

sns.boxplot(data=process, x='Operasi', y='Throughput per client', hue='Volume', palette='crest',ax=axes[0,0])
axes[0,0].set_title("Process: Throughput by Operation and File Size")

sns.boxplot(data=process, x='Operasi', y='Waktu total per client', hue='Volume', palette='crest',ax=axes[0,1])
axes[0,1].set_title("Process: Waktu by Operation and File Size")

sns.boxplot(data=thread, x='Operasi', y='Throughput per client', hue='Volume', palette='crest',ax=axes[1,0])
axes[1,0].set_title("Thread: Throughput by Operation and File Size")

sns.boxplot(data=thread, x='Operasi', y='Waktu total per client', hue='Volume', palette='crest',ax=axes[1,1])
axes[1,1].set_title("Thread: Waktu by Operation and File Size")

plt.tight_layout()
plt.show()
```
