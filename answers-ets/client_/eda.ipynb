```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('stress_test_log_process.csv')
```

```python
df.drop_duplicates()
```

```python
# Show occurence of each unique value in every column
for col in ['Operasi', 'Volume', 'Jumlah client worker pool', 'Jumlah server worker pool']:
    print(f"Unique values in {col}:")
    print(df[col].value_counts())
    print()
```

```python
df.sort_values(by=['Operasi', 'Volume', 'Jumlah client worker pool', 'Jumlah server worker pool'])
```

```python
sns.set(style="whitegrid")
sns.barplot(x='Jumlah client worker pool', y='Waktu total per client', data=df)
plt.ylabel('Waktu total per client (s)')
plt.xlabel('Jumlah Client')
```

```python
sns.set(style="whitegrid")
sns.barplot(x='Jumlah client worker pool', y='Throughput per client', data=df)
plt.ylabel('Throughput per client (ops/s)')
plt.xlabel('Jumlah Client')
```
